```python

import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

Grafica de facturacion por tienda.

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
facturacion = [facturacion1, facturacion2, facturacion3, facturacion4]
plt.bar(tiendas, facturacion, color='blue')
plt.title('Facturación por Tienda', fontweight="bold", fontsize= 19)
plt.ylabel("Facturaciónes")
plt.show()
```

Ingreso de informacion Latitud y longitud

```python
lat_all = pd.concat([
    tienda["lat"],
    tienda2["lat"],
    tienda3["lat"],
    tienda4["lat"]
], ignore_index=True)

lon_all = pd.concat([
    tienda["lon"],
    tienda2["lon"],
    tienda3["lon"],
    tienda4["lon"]
], ignore_index=True)
```

```python
plt.figure(figsize=(8, 6))

plt.hist2d(lon_all, lat_all, bins=50, cmap='hot')
plt.colorbar(label="Densidad de puntos")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Mapa de calor de ubicaciones de tiendas", fontweight="bold", fontsize= 13)
plt.show()
```
